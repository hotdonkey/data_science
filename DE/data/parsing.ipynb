```python
response = requests.get(
    'https://api.github.com/search/repositories',
    params={'q': 'requests+language:python'},
)
json_response = response.json()
# исследуем атрибут 'items' в ответе
repository = json_response['items'][0]
print(f'Repository name: {repository["name"]}')
```

```python
url1 = 'https://skillfactory.ru/'
url2 = 'https://habr.com/ru/companies/'
url3 = 'https://yandex.ru'
```

```python
USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"
headers = {"user-agent": USER_AGENT}
response1 = requests.get(url=url1, headers=headers)
response1.status_code
```

```python
urls = ['https://skillfactory.ru', 'https://habr.com/ru/companies/', 'https://yandex.ru']

USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"
headers = {"user-agent": USER_AGENT}

for url in urls:
    response = requests.get(url=url, headers=headers)
    print(response.status_code)
    print(response.text)
```
