```python
import pandas as pd
melb_df = pd.read_csv('data/melb_data_fe.csv')
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df['Quarters'] = melb_df['Date'].dt.quarter
melb_df['Quarters'].value_counts()
```

```python
melb_df = melb_df.drop('Quarters',axis=1)
display(melb_df.info())
```

```python
exclude_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
for column in melb_df.columns:
    if column not in exclude_list and melb_df[column].nunique() < 150:
        melb_df[column] = melb_df[column].astype('category')
```
