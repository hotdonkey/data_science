Вопрос 2. Проанализируйте динамику суммарного изменения дохода (сумму поля gross income) по дням на графике. Почему мы не можем использовать графическое представление изменения дохода по дням для анализа данных?

```python
shop_sum_dynamic = df.groupby(["City", "Date"])["gross income"].sum()
shop_sum_dynamic = pd.DataFrame(shop_sum_dynamic).reset_index()
```

```python
import matplotlib.pyplot as plt

# Создание графика
plt.figure(figsize=(12, 6))

# Перебор уникальных городов для построения линий на графике
for city in shop_sum_dynamic['City'].unique():
    subset = shop_sum_dynamic[shop_sum_dynamic['City'] == city]
    plt.plot(subset['Date'], subset['gross income'], label=city)

# Настройка графика
plt.title('Суммарный доход по городам по датам')
plt.xlabel('Дата')
plt.ylabel('Суммарный доход')
plt.legend(title='Город')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid()

# Показать график
plt.show()
```

Описывающий многоугольник — выпуклый многоугольник, вершинами которого являются точки диаграммы, при этом все остальные вершины находятся внутри многоугольника. На рисунке ниже вы видите пример построения описывающего многоугольника (показан синим цветом) и описывающего прямоугольника (показан красным цветом).

```python
ecars_df = pd.read_csv("./data/ecars.csv")
ecars_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import numpy as np

points = ecars_df[["PriceEuro", "TopSpeed_KmH"]].values


plt.scatter(points[:, 0], points[:, 1], color='blue')

# Вычисляем выпуклую оболочку
hull = ConvexHull(points)

# Рисуем выпуклую оболочку
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], color='red')

# Настройки графика
plt.title('Выпуклая оболочка для PriceEuro и TopSpeed')
plt.xlabel('PriceEuro')
plt.ylabel('TopSpeed_KmH')
plt.grid()
plt.show()
```

```python
# Определение координат точек
min_x = ecars_df["PriceEuro"].min()
max_x = ecars_df["PriceEuro"].max()
min_y = ecars_df["TopSpeed_KmH"].min()
max_y = ecars_df["TopSpeed_KmH"].max()

# Вычисление ширины и высоты
width = max_x - min_x
height = max_y - min_y

# Вычисление площади описывающего прямоугольника
area = width * height

print(f'Площадь описывающего прямоугольника: {area}')
```

### Задание 5
1.Используя модуль LabelEncoder из библиотеки sklearn, выполните кодирование категориального признака BodyStyle.

2.Найдите сумму полученных значений категориального признака и сохраните результат в переменную sum.

3.Найдите количество уникальных категорий полученного признака и сохраните результат в переменную cat_count.


```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
ecars_df['BodyStyle_encoded'] = label_encoder.fit_transform(ecars_df['BodyStyle'])


sum = ecars_df['BodyStyle_encoded'].sum()


cat_count = ecars_df['BodyStyle_encoded'].nunique()

# Вывод результатов
print(f'Сумма закодированных значений: {sum}')
print(f'Количество уникальных категорий: {cat_count}')
```
