### Линейные уравнения

$$3 x - 2 y -z = 15 $$
$$3 y + 10 z = 17$$
$$-5x+10y -20z=-75$$

```python
a=np.array([[3,-2,-1],
            [0,3,10],
            [-5,10,-20]])
b=np.array([15,17,-75])
```

```python
answer = np.linalg.solve(a,b)
```

```python
answer.astype("int")
```

```python
a @ answer
```

### Изображения

```python
from PIL import Image

image = Image.open("./data/numpy.jpeg")
display(image)
```

```python
image_array = np.array(image)
height, width, channels = image_array.shape
selected_part = image_array[:height // 2, :, :]
flipped_part = np.rot90(selected_part, 2)
image_array[:height // 2, :, :] = flipped_part
flipped_image = Image.fromarray(image_array)
display(flipped_image)
#flipped_image.save("./data/numpy_out.jpeg")
```
