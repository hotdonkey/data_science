```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Electric_Car.csv')
```

```python
df_pie = pd.DataFrame(
    df["Brand"].value_counts().reset_index())

top_5 = df_pie[:5]
rest = df_pie["count"].iloc[len(top_5):].sum()
rest = pd.DataFrame({"Brand":"Rest", "count":rest}, index=[0])
top_5 = pd.concat([top_5, rest]).reset_index(drop=True)
top_5["proportion"] = top_5["count"]/ top_5["count"].sum()



fig, ax = plt.subplots(figsize=(20, 15))
plt.pie(
    top_5["proportion"], 
    labels=top_5["Brand"], 
    autopct="%1.1f%%", 
    radius=3)

plt.savefig('auto_pie.png')

plt.show()
```
