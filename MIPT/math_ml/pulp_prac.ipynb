## Задание 6.1

Составьте оптимальный план перевозок со склада Nº1 и склада Nº2 в три торговых центра с учётом тарифов, запасов на складах и потребностей
торговых центров, которые указаны в таблице:

|   | ТЦ1 (110 шт.)  | ТЦ2 (150 шт.) |   ТЦ3 (140 шт.) |
|:------------- |:---------------:| -------------:|:-------------:|
| Склад Nº1 (180 шт.)    | 2 y. e.          | 5 y. e.       |3 у. е.        |
| Склад Nº2 (220 шт.)      | 7 y. e.       | 7 y. e.       |б у. е.     |



```python
problem = pulp.LpProblem("Logistic_Plan", pulp.LpMinimize)

# Переменные: количество отправляемых единиц от каждого склада в каждый торговый центр
x11 = pulp.LpVariable("x11", lowBound=0, cat=pulp.LpInteger)  # Склад Nº1 в ТЦ1
x12 = pulp.LpVariable("x12", lowBound=0, cat=pulp.LpInteger)  # Склад Nº1 в ТЦ2
x13 = pulp.LpVariable("x13", lowBound=0, cat=pulp.LpInteger)  # Склад Nº1 в ТЦ3

x21 = pulp.LpVariable("x21", lowBound=0, cat=pulp.LpInteger)  # Склад Nº2 в ТЦ1
x22 = pulp.LpVariable("x22", lowBound=0, cat=pulp.LpInteger)  # Склад Nº2 в ТЦ2
x23 = pulp.LpVariable("x23", lowBound=0, cat=pulp.LpInteger)  # Склад Nº2 в ТЦ3

# Целевая функция (минимизация затрат)
problem += 2 * x11 + 5 * x12 + 3 * x13 + 7 * x21 + 7 * x22 + 6 * x23

# Ограничения на потребности торговых центров
problem += x11 + x21 == 110  # ТЦ1
problem += x12 + x22 == 150  # ТЦ2
problem += x13 + x23 == 140  # ТЦ3

# Ограничения на запасы на складах
problem += x11 + x12 + x13 <= 180  # Склад Nº1
problem += x21 + x22 + x23 <= 220  # Склад Nº2


# Решаем
problem.solve()

# Результаты
print("Количество товаров от склада Nº1 в ТЦ1:", x11.varValue)
print("Количество товаров от склада Nº1 в ТЦ2:", x12.varValue)
print("Количество товаров от склада Nº1 в ТЦ3:", x13.varValue)
print("Количество товаров от склада Nº2 в ТЦ1:", x21.varValue)
print("Количество товаров от склада Nº2 в ТЦ2:", x22.varValue)
print("Количество товаров от склада Nº2 в ТЦ3:", x23.varValue)
print("Общие затраты на перевозку:", pulp.value(problem.objective))
```

## Задание 6.2

наименьшими.

| Работник/Задача | 1 | 2 | 3 | 4 | 5 |
|:------------:|:-----------:|:-----------:|:-----------:|:-----------:|:-----------:|
|      1     |      1000      |     12   |      10     |     19    |      8     |

|      3      |      10      |     3    |      1000     |      6     |      20    |

|      5     |      8    |     2    |      20     |      4     |      1000      |


```python
problem = pulp.LpProblem("Worker_dist", pulp.LpMinimize)

# Переменные
x11 = pulp.LpVariable("x11", lowBound=0, cat=pulp.LpBinary)
x12 = pulp.LpVariable("x12", lowBound=0, cat=pulp.LpBinary)
x13 = pulp.LpVariable("x13", lowBound=0, cat=pulp.LpBinary)
x14 = pulp.LpVariable("x14", lowBound=0, cat=pulp.LpBinary)
x15 = pulp.LpVariable("x15", lowBound=0, cat=pulp.LpBinary)

x21 = pulp.LpVariable("x21", lowBound=0, cat=pulp.LpBinary)
x22 = pulp.LpVariable("x22", lowBound=0, cat=pulp.LpBinary)
x23 = pulp.LpVariable("x23", lowBound=0, cat=pulp.LpBinary)
x24 = pulp.LpVariable("x24", lowBound=0, cat=pulp.LpBinary)
x25 = pulp.LpVariable("x25", lowBound=0, cat=pulp.LpBinary)

x31 = pulp.LpVariable("x31", lowBound=0, cat=pulp.LpBinary)
x32 = pulp.LpVariable("x32", lowBound=0, cat=pulp.LpBinary)
x33 = pulp.LpVariable("x33", lowBound=0, cat=pulp.LpBinary)
x34 = pulp.LpVariable("x34", lowBound=0, cat=pulp.LpBinary)
x35 = pulp.LpVariable("x35", lowBound=0, cat=pulp.LpBinary)

x41 = pulp.LpVariable("x41", lowBound=0, cat=pulp.LpBinary)
x42 = pulp.LpVariable("x42", lowBound=0, cat=pulp.LpBinary)
x43 = pulp.LpVariable("x43", lowBound=0, cat=pulp.LpBinary)
x44 = pulp.LpVariable("x44", lowBound=0, cat=pulp.LpBinary)
x45 = pulp.LpVariable("x45", lowBound=0, cat=pulp.LpBinary)

x51 = pulp.LpVariable("x51", lowBound=0, cat=pulp.LpBinary)
x52 = pulp.LpVariable("x52", lowBound=0, cat=pulp.LpBinary)
x53 = pulp.LpVariable("x53", lowBound=0, cat=pulp.LpBinary)
x54 = pulp.LpVariable("x54", lowBound=0, cat=pulp.LpBinary)
x55 = pulp.LpVariable("x55", lowBound=0, cat=pulp.LpBinary)

# Целевая функция

problem += 1000 * x11 + 12 * x12 + 10 * x13 + 19 * x14 + 8 * x15 +\
    12 * x21 + 1000 * x22 + 3 * x23 + 7 * x24 + 2 * x25 +\
        10 * x31 + 3 * x32 + 1000 * x33 + 6 * x34 +20 * x35 +\
        19 * x41 + 7 * x42 + 6 * x43 + 1000 * x44 + 4 * x45 +\
        8 * x51 + 2 * x52 + 20 * x53 + 4 * x54 + 1000 * x55
        
        
# Каждая задача должна быть выполнена ровно одним исполнителем
problem += x11 + x21 + x31 + x41 + x51 == 1  # Задача 1
problem += x12 + x22 + x32 + x42 + x52 == 1  # Задача 2
problem += x13 + x23 + x33 + x43 + x53 == 1  # Задача 3
problem += x14 + x24 + x34 + x44 + x54 == 1  # Задача 4
problem += x15 + x25 + x35 + x45 + x55 == 1  # Задача 5

# Каждый исполнитель может выполнять только одну задачу
problem += x11 + x12 + x13 + x14 + x15 <= 1  # Исполнитель 1
problem += x21 + x22 + x23 + x24 + x25 <= 1  # Исполнитель 2
problem += x31 + x32 + x33 + x34 + x35 <= 1  # Исполнитель 3
problem += x41 + x42 + x43 + x44 + x45 <= 1  # Исполнитель 4
problem += x51 + x52 + x53 + x54 + x55 <= 1  # Исполнитель 5

# Решаем
problem.solve()

# Вывод результатов
print("Статус:", pulp.LpStatus[problem.status])
for i in range(1, 6):
    for j in range(1, 6):
        if pulp.value(eval(f"x{i}{j}")) == 1:
            print(f"Рабочий {i} назначен на задачу {j}")

print("Минимальные затраты:", pulp.value(problem.objective))
```
