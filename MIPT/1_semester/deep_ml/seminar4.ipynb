
### Шаги теста Манна-Уитни

1. **Объединение данных**: 
   Объедините значения двух выборок и присвойте каждому значению ранг, начиная с 1 для наименьшего значения. Если значения совпадают, каждому из них присваивается средний ранг.

2. **Сумма рангов**:
   Рассчитайте сумму рангов для каждой из выборок:
   $$
   R_X = \sum_{i=1}^{n} \text{ранг}(X_i), \quad R_Y = \sum_{j=1}^{m} \text{ранг}(Y_j)
   $$
   где $ R_X $ и $ R_Y $ — суммы рангов для выборок $ X $ и $ Y $ соответственно.

3. **Вычисление статистики U**:
   Статистика Манна-Уитни $ U $ определяется для каждой группы следующим образом:
   $$
   U_X = n \cdot m + \frac{n (n + 1)}{2} - R_X
   $$
   $$
   U_Y = n \cdot m + \frac{m (m + 1)}{2} - R_Y
   $$
   где $ n $ и $ m $ — размеры выборок $ X $ и $ Y $ соответственно.

4. **Определение итоговой статистики U**:
   Итоговое значение $ U $ выбирается как минимальное из $ U_X $ и $ U_Y $:
   $$
   U = \min(U_X, U_Y)
   $$

5. **Определение p-значения**:
   Используя рассчитанное значение $ U $ и таблицу критических значений теста Манна-Уитни, определите, является ли различие значимым.


### Тест Манна-Уитни

```python
from scipy.stats import mannwhitneyu

# Данные для двух групп
group_1 = normal[feature]
group_2 = unnormal[feature]
# Проведение теста Манна-Уитни
stat, p_value = mannwhitneyu(group_1, group_2, alternative="two-sided")
# Вывод результата
print("U-статистика:", stat)
print("p-значение:", p_value)
# Интерпретация
alpha = 0.02  # порог значимости
if p_value < alpha:
    print("Отклоняем нулевую гипотезу: распределения отличаются.")
else:
    print("Не отклоняем нулевую гипотезу: распределения не отличаются.")
```

### P-test

### Формула t-статистики

$$
t = \frac{\bar{X} - \bar{Y}}{\sqrt{\frac{s_X^2}{n} + \frac{s_Y^2}{m}}}
$$

где:

- $\bar{X}$ и $\bar{Y}$ — средние значения выборок $X$ и $Y$,
- $s_X^2$ и $s_Y^2$ — выборочные дисперсии выборок $X$ и $Y$,
- $n$ и $m$ — размеры выборок.

### Степени свободы

Для приближенного расчёта степеней свободы используется формула:

$$
df = \frac{\left( \frac{s_X^2}{n} + \frac{s_Y^2}{m} \right)^2}{\frac{\left( \frac{s_X^2}{n} \right)^2}{n - 1} + \frac{\left( \frac{s_Y^2}{m} \right)^2}{m - 1}}
$$

где $df$ — число степеней свободы, используемое для нахождения p-значения из таблицы распределения Стьюдента.



```python
from scipy.stats import ttest_ind

# Данные для двух групп
group_1 = normal[feature]
group_2 = unnormal[feature]
# Выполнение t-теста для независимых выборок
t_stat, p_value = ttest_ind(group_1, group_2, alternative="two-sided")
# Вывод результата
print("t-статистика:", t_stat)
print("p-значение:", p_value)
# Интерпретация результата
alpha = 0.02  # порог значимости
if p_value < alpha:
    print("Отклоняем нулевую гипотезу: распределения  отличаются.")
else:
    print("Не отклоняем нулевую гипотезу: распределения  не отличаются.")
```
