```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/Cars.csv")
```

```python
df_bar = df["Brand"].value_counts(ascending=True)
fig, ax = plt.subplots(figsize=(20,15))

plt.barh(df_bar.index, df_bar.values)
plt.title("   ")

plt.savefig('./data/auto_bar.png')
plt.show()
```
