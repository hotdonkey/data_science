```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from stl import mesh

# 1. Параметры цилиндра
N = 20  # количество угловых секторов
height = 1  # высота цилиндра
radius = 1  # радиус основания

# 2. Сформируем вершины для основания
theta = np.linspace(0, 2 * np.pi, N, endpoint=False)  # углы
x1 = radius * np.cos(theta)  # координаты x первого основания
y1 = radius * np.sin(theta)  # координаты y первого основания
z1 = np.zeros(N)  # координаты z первого основания (на уровне 0)

# 3. Сформируем вершины второго основания
x2 = radius * np.cos(theta)  # координаты x второго основания (на уровне 1)
y2 = radius * np.sin(theta)  # координаты y второго основания (на уровне 1)
z2 = np.ones(N)  # координаты z второго основания (на уровне 1)

# 4. Объединяем координаты вершин
vertices = np.array([
    [x1[i], y1[i], z1[i]] for i in range(N)
] + [
    [x2[i], y2[i], z2[i]] for i in range(N)
])

# 5. Создаем грани
faces = []
for i in range(N):
    # Создаем две треугольные грани для каждого бокового сегмента
    faces.append([i, (i + 1) % N, N + (i + 1) % N])  # первая грань
    faces.append([i, N + (i + 1) % N, N + i])        # вторая грань

faces = np.array(faces)

# 6. Создаем STL объект
cylinder = mesh.Mesh(np.zeros(faces.shape[0], dtype=mesh.Mesh.dtype))
for i, f in enumerate(faces):
    for j in range(3):
        cylinder.vectors[i][j] = vertices[f[j], :]

# 7. Визуализируем цилиндр
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.add_collection3d(mplot3d.art3d.Patch3DCollection(cylinder.vectors))

# 8. Настройка отображения
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Цилиндр')
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([0, 1.5])

plt.show()

# 9. Сохранение в формате STL
cylinder.save('cylinder.stl')
```
