## Altair

https://altair-viz.github.io/index.html

```python
import altair as alt
alt.themes.enable('dark')
alt.data_transformers.disable_max_rows();
alt.warnings.simplefilter(action='ignore')
```

```python
chart = alt.Chart(df).mark_point().encode(x='make',y='price').interactive()
chart
```

```python
chart = alt.Chart(df).mark_bar().encode(x='make',y='sum(price)').interactive()
chart
```

```python
selection = alt.selection_multi(fields=['horsepower'],bind='legend')

chart = alt.Chart(df).mark_point().add_selection(selection).encode(
x='make',
y='price',color='horsepower:N',
opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).interactive()
chart
```

```python
# Определение выбора через прямоугольную область
brush = alt.selection_interval(encodings=['x','y'])

# Создание диаграммы рассеяния с выбором через прямоугольник
chart = alt.Chart(df).mark_circle(size=100).encode(
    x=alt.X('make', title='Бренд'),
    y=alt.Y('price', title='Цена'),
    color=alt.Color('horsepower', title='Мощность'),
    opacity=alt.condition(brush, alt.value(1), alt.value(0.2)),

).add_selection(
    brush
).properties(
    title='Диаграмма рассеяния цен'
)
chart
```

## Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
g = sns.barplot(
data=df,
x="make",
y="price",
palette="dark",
errorbar=None # отключаем доверительный интервал
)
plt.xticks(rotation=90)
```

```python
df_=df[df.describe().columns]
df_
```

```python
corr_matrix = df_.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix,
            annot=True,       # Отображение значений корреляции
            fmt=".2f",        # Формат чисел (2 знака после запятой)
            cmap='coolwarm',  # Цветовая палитра
            linewidths=0.5,   # Ширина линий между клетками
            linecolor='white',# Цвет линий между клетками
            square=True,      # Квадратные клетки
            cbar_kws={"shrink": 0.75})  # Параметры цветовой шкалы
```
