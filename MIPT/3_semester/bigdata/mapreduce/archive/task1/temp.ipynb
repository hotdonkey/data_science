```python
def timecounter(data_string: dict):
    # Вытащим id и список с рабочими днями
    id = data_string["business_id"]
    hours = data_string["hours"]

    minutes = list()
    
    # подсчитаем суммарное колличество рабочих минут
    try:
        for i in hours.values():
            working_time = i.split("-")
            time_start = datetime.strptime(working_time[0], '%H:%M').time()
            time_end = datetime.strptime(working_time[1], '%H:%M').time()
            
            total_start = time_start.hour * 60 + time_start.minute
            total_end = time_end.hour * 60 + time_end.minute
            
            if total_end <= total_start:
                # Работа после полуночи
                time_delta = (24 * 60 - total_start) + total_end
            else:
                time_delta = total_end - total_start

            
            minutes.append(time_delta)
            
    # если null вернем ноль
    except AttributeError as e:
        minutes = [0]
    
    # возвращаем строку id+суммарное колличество рабочих минут
    return f"{id}\t{sum(minutes)}"
    
```

```python

def timecounter(business_id, hours):
    total_minutes = 0
    
    # Если hours отсутствует или None
    if not hours:
        return f"{business_id}\t0"
    
    try:
        for day_hours in hours.values():
            # Проверяем на "0:0-0:0" - выходной день
            if day_hours == "0:0-0:0":
                continue
                
            working_time = day_hours.split("-")
            if len(working_time) != 2:
                continue
                
            time_start_str = working_time[0].strip()
            time_end_str = working_time[1].strip()
            
            # Парсим время
            time_start = datetime.strptime(time_start_str, '%H:%M').time()
            time_end = datetime.strptime(time_end_str, '%H:%M').time()
            
            total_start = time_start.hour * 60 + time_start.minute
            total_end = time_end.hour * 60 + time_end.minute
            
            if total_end < total_start:
                # Работа после полуночи
                time_delta = (24 * 60 - total_start) + total_end
            else:
                time_delta = total_end - total_start
                
            total_minutes += time_delta
            
    except (ValueError, AttributeError, KeyError) as e:
        # В случае ошибок парсинга считаем 0 для этого бизнеса
        return f"{business_id}\t0"
    
    return f"{business_id}\t{total_minutes}"
```
