```python
import ast
import json

def dict_to_objects_string(data):
    """
    Преобразует словарь в строку формата ключ:значение,
    а вложенные словари выводятся как ключ:подключ:значение
    """
    result_lines = []
    
    for key, value in data.items():
        # Если значение строка и похожа на словарь
        if isinstance(value, str) and value.startswith('{') and value.endswith('}'):
            try:
                # Парсим строку как словарь
                inner_dict = ast.literal_eval(value)
                # Для каждого элемента вложенного словаря
                for inner_key, inner_value in inner_dict.items():
                    # Добавляем строку ключ:вложенный_ключ:вложенное_значение
                    result_lines.append(f"{key}:{inner_key}:{inner_value}")
            except:
                # Если не удалось распарсить, выводим как есть
                result_lines.append(f"{key}:{value}")
        else:
            # Простое значение
            result_lines.append(f"{key}:{value}")
    
    return ":".join(result_lines)

input_dict = test[10]

transformed_dict = dict_to_objects_string(input_dict)

transformed_dict
```

```python
import ast

test = {
    "OutdoorSeating": "False",
    "BusinessParking": "{'garage': False, 'street': True, 'validated': True, 'lot': True, 'valet': False}",
    "WiFi": "u'free'",
    "RestaurantsPriceRange2": "2",
    "HasTV": "False",
    "Alcohol": "u'none'",
    "RestaurantsTakeOut": "True",
    "BikeParking": "True",
    "Smoking": "u'no'",
    "Music": "{'dj': False, 'background_music': False, 'no_music': False, 'jukebox': False, 'live': True, 'video': False, 'karaoke': False}",
    "Caters": "False",
    "DogsAllowed": "False",
    "CoatCheck": "False",
    "WheelchairAccessible": "True"
}

result = {}

for key, val_str in test.items():
    try:
        # Пытаемся распарсить как Python-объект
        parsed = ast.literal_eval(val_str)
    except (ValueError, SyntaxError):
        # Если не получилось оставляем как строку
        parsed = val_str

    if isinstance(parsed, dict):
        # Это вложенный словарь раскрываем
        for subkey, subval in parsed.items():
            result[f"{key}:{subkey}"] = subval
    else:
        # Простое значение (bool, str, int и т.д.)
        result[key] = parsed

# Вывод
for k, v in sorted(result.items()):
    print(f"{k}:{v}")
```

```python
def parser(string):
    result = {}

    for key, val_str in string.items():
        try:
            # Пытаемся распарсить как Python-объект
            parsed = ast.literal_eval(val_str)
        except (ValueError, SyntaxError):
            # Если не получилось оставляем как строку
            parsed = val_str

        if isinstance(parsed, dict):
            # Это вложенный словарь раскрываем
            for subkey, subval in parsed.items():
                result[f"{key}:{subkey}"] = subval
        else:
            # Простое значение (bool, str, int и т.д.)
            result[key] = parsed

    # Вывод
    for k, v in sorted(result.items()):
        print(f"{k}:{v}")
    
```
