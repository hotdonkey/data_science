```python
customers_new = customers.drop(columns=["customer_unique_id"])
```

---
###  orders, order_payments, order_reviews, order_items  order_id

```python
orders_new = pd.merge(left=orders, right=order_payments,
                right_on="order_id", left_on="order_id", how="left")
orders_new.head()
```

```python
orders_new = pd.merge(
    left=orders_new, right=order_reviews,
    right_on="order_id", left_on="order_id", how="left")
orders_new.head()
```

```python
orders_new = pd.merge(
    left=orders_new, right=orders_items,
    right_on="order_id", left_on="order_id", how="left")
orders_new.head()
```

---
###  products  prod_cat.._name...

```python
products_new = pd.merge(
    left=products, right=product_category_name_translation,
    right_on="product_category_name", left_on="product_category_name", how="left")
products_new.head()
```

---
###  orders_new  sellers  seller_id

```python
orders_new = pd.merge(
    left=orders_new, right=sellers,
    right_on="seller_id", left_on="seller_id", how="left")
orders_new.head()
```

---
###  orders_new  products_new

```python
orders_new = pd.merge(
    left=orders_new, right=products_new,
    right_on="product_id", left_on="product_id", how="left")

orders_new.head()
```

---
###   orders_new  customers

```python
orders_clust = pd.merge(
    left=orders_new, right=customers_new,
    right_on="customer_id", left_on="customer_id", how="left")
orders_clust.head()
```

```python
orders_clust.to_csv("./dataset/joined_df.csv", index=False)
```

```python
#   id
orders_for_clust = orders_clust.drop(
    columns=["order_id", "review_id", "product_id", "seller_id"])
```
