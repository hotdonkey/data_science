### Оформим все в один пайплайн

```python
# Загрузка данных
def load_data():
    customers = pd.read_csv("./clean_data/customers.csv")
    geolocation = pd.read_csv("./clean_data/geolocation.csv")
    order_pay = pd.read_csv("./clean_data/order_payments.csv")
    reviews = pd.read_csv("./clean_data/order_reviews.csv")
    orders = pd.read_csv("./clean_data/orders.csv")
    item = pd.read_csv("./clean_data/orders_items.csv")
    category_name = pd.read_csv(
        "./clean_data/product_category_name_translation.csv")
    products = pd.read_csv("./clean_data/products.csv")
    sellers = pd.read_csv("./clean_data/sellers.csv")
    return customers, geolocation, order_pay, reviews, orders, item, category_name, products, sellers


# Объединение данных
def merge_data(orders, item, order_pay, reviews, products, customers, sellers, category_name):
    df = orders.merge(item, on='order_id', how='left')
    df = df.merge(order_pay, on='order_id', how='outer', validate='m:m')
    df = df.merge(reviews, on='order_id', how='outer')
    df = df.merge(products, on='product_id', how='outer')
    df = df.merge(customers, on='customer_id', how='outer')
    df = df.merge(sellers, on='seller_id', how='outer')
    df = df.merge(category_name, on="product_category_name", how="left")
    return df


# Очистка данных: удаление строк без customer_id
def filter_customers(df):
    return df[~df["customer_id"].isna()]


# Создание временных признаков
def create_time_features(df):
    time_features = [
        "order_purchase_timestamp", "order_approved_at",
        "order_delivered_carrier_date", "order_delivered_customer_date",
        "order_estimated_delivery_date", "shipping_limit_date",
        "review_creation_date", "review_answer_timestamp"
    ]
    for col in time_features:
        df[col] = pd.to_datetime(df[col], errors='coerce')

    df_time = df[["customer_id", *time_features]].copy()

    # Вычисление временных интервалов
    df_time["time_btw_purch_approve"] = df_time["order_approved_at"] - \
        df_time["order_purchase_timestamp"]
    df_time["time_btw_approval_delivery_to_carrier"] = df_time["order_delivered_carrier_date"] - \
        df_time["order_approved_at"]
    df_time["time_btw_carrier_customer"] = df_time["order_delivered_customer_date"] - \
        df_time["order_delivered_carrier_date"]
    df_time["time_btw_customer_est_date"] = df_time["order_estimated_delivery_date"] - \
        df_time["order_delivered_customer_date"]
    df_time["time_btw_est_date_limit_date"] = df_time["shipping_limit_date"] - \
        df_time["order_estimated_delivery_date"]
    df_time["time_btw_customer_rewiew"] = df_time["review_creation_date"] - \
        df_time["order_delivered_customer_date"]
    df_time["time_btw_rewiew_answer"] = df_time["review_answer_timestamp"] - \
        df_time["review_creation_date"]

    # Удаление исходных временных колонок
    df_time = df_time.drop(columns=time_features)
    
    columns_to_transform = df_time.columns.to_list()[1:]
    df_time = df_time.groupby("customer_id")[columns_to_transform].mean()
    columns_to_transform = df_time.columns.to_list()
    
    for column in columns_to_transform:
        df_time[column] = df_time[column].apply(
            lambda x: pd.Timedelta(x).total_seconds())
    
    return df_time


# Агрегация данных
def aggregate_data(df):
    # Количество заказов на одного клиента
    cust_clust = df.groupby("customer_id")["order_id"].count().reset_index()
    cust_clust = cust_clust.rename(columns={"order_id": "number_of_orders"})

    # Средние значения для числовых колонок
    mean_values = df.groupby("customer_id").agg({
        "price": "mean",
        "freight_value": "mean",
        "review_score": "mean",
    })

    # Общая сумма транзакций
    customer_transactions = df.groupby(
        "customer_id")["payment_value"].sum().reset_index()
    customer_transactions = customer_transactions.rename(
        columns={"payment_value": "total_spent"})

    # Мода для категориальных колонок
    categories = df.groupby("customer_id").agg({
        "product_category_name_english": lambda x: x.mode().iloc[0] if not x.mode().empty else None,
        "payment_type": lambda x: x.mode().iloc[0] if not x.mode().empty else None,
    })

    # Бинарный признак для payment_installments
    payment_installments = df[["customer_id", "payment_installments"]]
    payment_installments["payment_installments"] = payment_installments["payment_installments"].apply(
        lambda x: 1 if x > 1 else 0)
    payment_installments = payment_installments.drop_duplicates()

    # Город клиента
    location = df[["customer_id", "customer_city"]]
    location = location.drop_duplicates().dropna()

    # Объединение всех агрегированных данных
    cust_clust = cust_clust.merge(mean_values, on="customer_id", how="left")
    cust_clust = cust_clust.merge(
        customer_transactions, on="customer_id", how="left")
    cust_clust = cust_clust.merge(categories, on="customer_id", how="left")
    cust_clust = cust_clust.merge(
        payment_installments, on="customer_id", how="left")
    cust_clust = cust_clust.merge(location, on="customer_id", how="left")

    # Удаление строк с пропущенными значениями
    cust_clust = cust_clust.dropna()
    return cust_clust


# Главная функция пайплайна
def main_pipeline():
    # Шаг 1: Загрузка данных
    customers, geolocation, order_pay, reviews, orders, item, category_name, products, sellers = load_data()

    # Шаг 2: Объединение данных
    df = merge_data(orders, item, order_pay, reviews, products,
                    customers, sellers, category_name)

    # Шаг 3: Фильтрация данных
    df = filter_customers(df)

    # Шаг 4: Создание временных признаков
    df_time = create_time_features(df)

    # Шаг 5: Агрегация данных
    cust_clust = aggregate_data(df)
    
    final_data = cust_clust.merge(df_time, on="customer_id", how="left")

    return final_data


# Запуск пайплайна
final_dataset = main_pipeline()
```

```python
final_dataset.to_csv("./dataset/clusterisation_dataset.csv", index=False)
```
