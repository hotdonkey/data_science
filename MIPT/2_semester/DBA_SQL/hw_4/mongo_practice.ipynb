```python
#Начать выполнение задания тут
import pymongo
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
credentials = pd.read_csv(
    "../credentials.txt").to_string().split()[3].replace("[", "").replace("]", "")

client = pymongo.MongoClient(credentials)
# Вывести список БД для проверки подключения
print(client.list_database_names())
```

```python
data = pd.read_excel("./data/data_for_mongodb.xlsx", engine='openpyxl')
data_dict = data.to_dict(orient='records')
data_dict
```

```python
db = client['lections']
```

```python
db.create_collection("users")
```

```python
db.list_collection_names()
```

```python
db.users.insert_many(data_dict)
```

```python
list(db.users.find())
```

```python
db.users.update_one({"Student Name": "Alexandra"}, {
                    "$set": {"Student Name": "Jorick"}})
```

```python
list(db.users.find())
```

```python
list(db.users.find({"Student Name": "Jorick"}))
```
