### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу `Некорректный ввод`.

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
[42, 52, 14]

```

**Пример №2:**
```
Ввод:
108, 138, 42, 52, 14
189, 202, 249, 24, 15, 3

Вывод:
[]

```

**Пример №3:**
```
Ввод:
108, 138, 42, 52, четырнадцать
189, 202, 249, 24, 15, 3

Вывод:
Некорректный ввод

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')
# ваш код здесь

# Основная функция
def pure_intersection(seq1, seq2):
    """
    Проверка пересечения последовательностей
    пользовательского ввода идентификаторов
    товаров и услуг

    Args:
        seq1 (str): 1 последовательность для сравнения
        (формат ввода: запятая + пробел)

        seq2 (str): 2 последовательность для сравнения
        (формат ввода: запятая + пробел)
    """
    # Конвертация пользовательского ввода в формат списков
    lst1 = [i for i in seq1.replace(",", " ").replace(".", " ").split()]
    lst2 = [i for i in seq2.replace(",", " ").replace(".", " ").split()]

    # Проверка формата данных
    try:
        for element in enumerate(lst1):
            lst1[element[0]] = int(lst1[element[0]])
        for element in enumerate(lst2):
            lst2[element[0]] = int(lst2[element[0]])

    except ValueError:
        print("Некорректный ввод")
        return "Некорректный ввод"

    # Преобразование списков во множества
    set1 = set(lst1)
    set2 = set(lst2)

    # Отображение итога
    print(list(set1.intersection(set2)))

    # Возврат значения для дальнейшего использования
    return list(set1.intersection(set2))


# Напишем тест для функции test_pure_intersection
def test_pure_intersection():
    '''Простой тест по предоставленным тестовым данным'''

    print("____Начало теста______")
    assert pure_intersection(
        "108, 138, 42, 52, 14", "109, 13, 52, 32, 42, 14, 109"
    ) == [42, 52, 14]

    assert pure_intersection("108, 138, 42, 52, 14",
                             "189, 202, 249, 24, 15, 3") == []
    assert (
        pure_intersection(
            '108, 138, 42, 52, "четырнадцать"', "189, 202, 249, 24, 15, 3"
        )
        == "Некорректный ввод"
    )
    print("____Конец теста______")


# Запуск теста
test_pure_intersection()

# Запуск функции
pure_intersection(input_string1, input_string2)
````

## Задание №2

#### Найти минимум и максимум

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию `input()`).

Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять») — в этом случае слово надо игнорировать.

**Ваша задача** — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.

Для нахождения минимума и максимума напишите функцию `find_min_max`, которая будет возвращать два числа.

**Пример:**

```
Ввод:
-2,56 25,002 восемь 19,13 13 -37,5

Вывод:
Minimum: -37.5
Maximim: 25.002
```



**Подсказка №1:**

Для проверки того, состоит ли строка `s` только из букв, можно воспользоваться методом `s.isalpha()`.

**Подсказка №2:**

Для поиска минимума и максимума в списке можно воспользоваться встроенными функциями `min()` и `max()`.

**Критерии оценивания:**

- Задача решена корректно, учтены все условия (**2 балла**):
    - программа корректно обрабатывает ввод пользователя и работает на любых входных данных;
    - программа корректно обрабатывает присутствие слов во вводе пользователя;
    - программа корректно рассчитывает минимум и максимум числовой последовательности;
    - решение сопровождено комментариями.
- Задача решена корректно, но не выполнено одно из условий (**1 балл**):
    - программа некорректно обрабатывает ввод пользователя и работает только на конкретном примере входных данных;
    - программа некорректно обрабатывает присутствие слов во вводе пользователя, что приводит к ошибкам при выполнении программы;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): программа некорректно рассчитывает минимум и максимум числовой последовательности.

````python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь

# Основная функция
def find_min_max(seq):
    """
    Нахождение минимума и максимума последовательности

    Args:
        seq (str): 1 последовательность значений
        (формат ввода: пробел)

    Return:
        Кортеж минимального и максимального значения
    """
    # Делаем влючение по условию возможности конвертации
    # с использованием пайплайна предобработки
    lst_from_seq = [float(i) for i in seq.replace(",", ".")\
        .split() if not i.isalpha()]

    # Находим значения с использованием встроенных функций
    # (встроенное быстрее самописного в этом случае будет)
    min_value = min(lst_from_seq)
    max_value = max(lst_from_seq)

    # Выводим результат в консоль
    print(f"Минимальное значение: {min_value}, максимальное значение: {max_value}")

    # Возвращаем значение для последующего использования (в частности для теста)
    return min_value, max_value

# Тест-функция
def test_find_min_max():
    print("____Начало теста_____")
    assert find_min_max("-2,56 25,002 восемь 19,13 13 -37,5") == (-37.5, 25.002)

    assert find_min_max("-2356 25 восемь 19,13 13 -37,5") == (-2356, 25)

    print(True)
    print("____Конец теста_____")


# Запуск теста
test_find_min_max()

# Запуск функции
find_min_max(input_string)
````

## Задание №3

#### Найти медиану

**Компания:** «Авито»

**Уровень:** Лёгкий

---
Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---
Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел неупорядоченную последовательность чисел.

Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.

Для поиска медианы в списке напишите функцию `find_median()`, которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.

Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза `Некорректный ввод`.

**Пример №1:**
```
Ввод:
1, 5, 2, 3, 6

Вывод:
Median: 3.0
```
```
Ввод:
100, 5, 2, 4, 3, 6

Вывод:
Median: 4.5
```
```
Ввод:

Вывод:
Некорректный ввод
```
```
Ввод:
десять, 10, пять, 7, семь

Вывод:
Некорректный ввод
```


**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - ввод пользователя конвертируется в список из чисел;
    - функция для расчёта медианы работает корректно для списка чётной и нечётной длины;
    - в случае некорректного ввода функция возвращает `None`;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для расчёта медианы некорректно работает для списка чётной или нечётной длины;
    - в случае некорректного ввода функция отрабатывает с ошибкой;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм расчёта медианы для обоих случаев, результат работы программы зависит от вводимой последовательности чисел и т. д.

````python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь

# Основная функция
def find_median(seq):
    """Поиск медианы в неупорядоченной последовательности чисел.

    Args:
        seq (str): пользовательская последовательность

    Return:
        Числовое значение медианы
    """
    try:
        # Конвертируем последовательность
        lst_from_seq = [float(i) for i in seq.replace(",", " ").split()]

        # Сортируем на месте
        lst_from_seq.sort()

        # Пробегаем по логике количества элементов в последовательности (чет/нечет)
        if len(lst_from_seq) % 2 != 0:
            result = lst_from_seq[int(len(lst_from_seq) / 2)]
            print(f'Медиана: {result}')
            return result
        else:
            result = (
                lst_from_seq[int(len(lst_from_seq) / 2 - 0.5)]  # Сдвигаем индексы
                + lst_from_seq[int(len(lst_from_seq) / 2 + 0.5)]
            ) / 2
            print(f'Медиана: {result}')
            return result
    except (IndexError, ValueError) as e:
        print(f"Некорректный ввод: {e}")
        return None

# Тест-функция
def test_find_median():
    """Тест"""
    print("____Начало теста_____")
    assert find_median("1, 5, 2, 3, 6") == 3

    assert find_median("100, 5, 2, 4, 3, 6") == 4.5

    assert find_median("") == None

    assert find_median("десять, 10, пять, 7, семь") == None

    print("Тест пройден")
    print("____Конец теста_____")


# Запуск теста
test_find_median()

# Запуск функции
find_median(input_string)
````

````python
input_string = input('Введите число словами: ')
# ваш код здесь

# Основная функция
def transform_string_to_integer(seq):
    """Функция перевода прописи в числовое значение

    Args:
        seq (str): входная последовательность

    Returns:
        int: Число
    """
    # Словарь, отображающий числовые значения слов
    numbers = {
        'ноль': 0, 'один': 1, 'две': 2, 'два': 2, 'три': 3, 'четыре': 4, 'пять': 5,
        'шесть': 6, 'семь': 7, 'восемь': 8, 'девять': 9, 'десять': 10,
        'одиннадцать': 11, 'двенадцать': 12, 'тринадцать': 13, 'четырнадцать': 14,
        'пятнадцать': 15, 'шестнадцать': 16, 'семнадцать': 17, 'восемнадцать': 18,
        'девятнадцать': 19, 'двадцать': 20, 'тридцать': 30, 'сорок': 40,
        'пятьдесят': 50, 'шестьдесят': 60, 'семьдесят': 70, 'восемьдесят': 80,
        'девяносто': 90, 'сто': 100, 'двести': 200, 'триста': 300, 'четыреста': 400,
        'пятьсот': 500, 'шестьсот': 600, 'семьсот': 700, 'восемьсот': 800, 'девятьсот': 900,
        'тысяча': 1000, 'миллион': 1000000, 'миллиард': 1000000000
    }

    # Вспомогательная функция для поиска наиболее близкого значения в словаре
    # рассчет расстояния Левенштейна (спасибо вики)
    def find_similar_key(input_key, dictionary=numbers):
        """Функция рассчета расстояния Левенштейна

        Args:
            input_key (str): ключ который не нашли
            dictionary (dict): словарь сопоставления значений

        Returns:
            Число
        """
        # Инициируем растояние бесконечностью
        min_distance = float('inf')
        # Инициируем найденый ключ (пока нулем)
        closest_key = None
        # Пройдем циклом по генератору (экономя память)
        for key in dictionary.keys():
            # Расчитаем суммарное расстояние по кортежам состоящим из
            # пар букв из имеющегося словаря и вошедшего ключа, присваивая каждому
            # непарному кортежу значение 1 и затем суммируя.
            # Так как zip реконфигурирует итоговый размер 2 входящих
            # последовательностей по меньшей избавляем себя от проблем с окончаниями.
            distance = sum(1 for a, b in zip(input_key, key) if a != b)
            # Обновим минимальное расстояние и имитируем градиентный спуск
            if distance < min_distance:
                min_distance = distance
                closest_key = key
        return closest_key


    words = seq.split()  # Разбиваем входную строку на слова
    result = 0 # Инициируем результат нулем
    for w in words: # Бежим по элементам строки, обращаясь к словарю
        if w in numbers:
            result += numbers[w] # Если находим, то скидываем в результат
        else:
            result *= numbers[find_similar_key(w)] # Если нет, то полагаем ключ разрядом, ищем его и умножаем

    print(f"Результирующее значение: {result}")
    return result

# Тест-функция
def test_transform_string_to_integer():
    """Тест"""
    print("____Начало теста_____")
    assert transform_string_to_integer("один") == 1

    assert transform_string_to_integer("двадцать") == 20

    assert transform_string_to_integer("двести сорок шесть") == 246

    assert transform_string_to_integer("семьсот восемьдесят три тысячи девятьсот девятнадцать") == 783919

    assert transform_string_to_integer("три миллиона два") == 3000002

    print("Тест пройден")
    print("____Конец теста_____")

# Запуск теста
test_transform_string_to_integer()

# Запуск функции
transform_string_to_integer(input_string)
````

#### Разложить факториал на простые множители

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Пользователь с клавиатуры (через функцию `input()`) вводит некоторое целое число `n`.

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

---

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

Например, факториал числа $3$ будет равен: $3! = 1 * 2 * 3 = 6$.

---

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

**Пример №1:**
```
Ввод:
5

Вывод:
5! = 2^3 * 3 * 5
```

**Пример №2:**
```
Ввод:
22

Вывод:
22! = 2^19 * 3^9 * 5^4 * 7^3 * 11^2 * 13 * 17 * 19
```

**Пример №3:**
```
Ввод:
25

Вывод:
25! = 2^22 * 3^10 * 5^6 * 7^3 * 11^2 * 13 * 17 * 19 * 23

```



**Подсказка №1:**

**Подсказка №2:**

Алгоритм нахождения простых множителей числа вы можете найти [здесь](https://foxford.ru/wiki/informatika/razlozhenie-chisla-na-mnozhiteli-v-python).

````python
n = input("Введите число: ")
# ваш код здесь


# Основная функция
def decompose_factorial(number):
    """Принимает целое число и вычисляет его факториал, затем разлагает факториал на простые множители.

    Args:
        number (int): Целое число, для которого нужно вычислить факториал и разложить его на множители.

    Returns:
        str: Строка, представляющая разложение факториала на простые множители.
    """

    def factorial_eq(number):
        """Вычисляет факториал заданного числа.

        Args:
            number (int): Число, для которого нужно вычислить факториал.

        Returns:
            int: Значение факториала заданного числа.
        """
        if number == 1 or number == 0:
            return 1
        return number * factorial_eq(number - 1)

    def factorize_number(factorial):
        """Разлагает число на простые множители.

        Args:
            factorial (int): Число, которое нужно разложить на множители.

        Returns:
            List[int]: Список простых множителей заданного числа.
        """
        factors = []  # список для хранения множителей
        divisor = 2  # начинаем с делителя 2
        while factorial > 1:
            while (
                factorial % divisor == 0
            ):  # если число делится на делитель без остатка

                factors.append(divisor)  # добавляем делитель в список множителей

                factorial //= divisor  # делим число на делитель

            divisor += 1  # увеличиваем делитель
        return factors

    # Для простоты и робастности
    number = int(number)

    # Cразу срежем углы на 0 и 1:
    if number == 0 or number == 1:
        print(result := f"{number}! = 1")
        return result
    else:
        # Расчитаем факториал
        factorial_result = factorial_eq(number)
        # Найдем все мнодители и поместим в список
        factorialised_list = factorize_number(factorial_result)
        # На основе списка выведем словарь содержащий степени множителей
        fac_dict = {k: 0 for k in set(factorialised_list)}
        # Заполним словарь количеством вхождений элементов == показатель степени
        for i in factorialised_list:
            fac_dict[i] += 1

        # Начинаем склеивать строку ответа используя списковое включение по итемам словаря
        # напихав валрусов забавы ради
        print("Ответ:",
              result:=
              f"{number}! = {' * '.join([f'{k}^{v}' if v > 1 else str(k) for k, v in fac_dict.items()])}")

        return result


# Тест-функция
def test_decompose_factorial():
    """Тест"""
    print("____Начало теста_____")

    assert decompose_factorial("5") == "5! = 2^3 * 3 * 5"
    assert (
        decompose_factorial("22")
        == "22! = 2^19 * 3^9 * 5^4 * 7^3 * 11^2 * 13 * 17 * 19"
    )
    assert (
        decompose_factorial("25")
        == "25! = 2^22 * 3^10 * 5^6 * 7^3 * 11^2 * 13 * 17 * 19 * 23"
    )

    print("Тест пройден")
    print("____Конец теста_____")


# Запуск теста
test_decompose_factorial()

# Запуск функции
decompose_factorial(n)
````

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:
zzzab

step 1: zaba
step 2: zbb
step 3: zc

output:
cz
```

```
input:
hhakafh

step 1: haakfi
step 2: hkfib

output:
hkfib
```

**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)
````

````python
# ваш код здесь
s = input('Введите строку: ')

def replace_duplicates(s):
    """
    Заменяет дубликаты символов в строке на следующие символы в алфавите и возвращает измененную строку.

    Args:
    s (str): Входная строка

    Returns:
    str: Измененная строка без дубликатов
    """

    # Встроенный алфавит
    alphabet = "abcdefghijklmnopqrstuvwxyz"

    # Подготовка ввода для обработки
    test_lst = []
    for i in s:
        test_lst.append(alphabet.find(i))  # Преобразование символов в индексы алфавита

    # Вспомогательная функция для поиска дубликатов
    def check_duplicates(test_lst=test_lst) -> dict:
        duplicates = {}
        for i, v in enumerate(test_lst):
            if v in duplicates:
                duplicates[v].append(i)
            else:
                duplicates[v] = [i]

        # Оставляем только дубликаты
        duplicates = {k: v for k, v in duplicates.items() if len(v) > 1}

        return duplicates

    # Инициализация словаря дубликатов
    duplicates = check_duplicates()

    # Начинаем крутить по списку выискивая и подменяя дубликаты
    counter = 1
    while True:
        # Первый дубликат
        dup_list = duplicates[list(duplicates.keys())[0]]
        a = dup_list[0]
        b = dup_list[1]

        test_lst.pop(b)  # Удаляем второй дубликат
        test_lst[a] += 1  # Увеличиваем первый дубликат на 1

        # Если вышли за пределы алфавита, возвращаемся к началу
        if (test_lst[a] > len(alphabet) - 1):
            test_lst[a] = 0


        # Преобразование индексов обратно в символы
        result = ""
        for i in test_lst:
            result += alphabet[i]

        print(f"Step {counter}: {result}")

        counter += 1

        duplicates = check_duplicates()  # Проверяем остались ли дубликаты

        # Как исчерпали все - выходим
        if len(duplicates) == 0:
            break

    return sorted([i for i in result]) # Чтоб проверить тестами


def test_replace_duplicates():
    """Тест"""
    print("____Начало теста_____")

    assert replace_duplicates("zzzab") == sorted([i for i in "cz"])
    assert replace_duplicates("hhakafh") == sorted([i for i in "hkfib"])

    print("Тест пройден")
    print("____Конец теста_____")


# Запуск теста
test_replace_duplicates()

# Запуск функции
replace_duplicates(s)
````
